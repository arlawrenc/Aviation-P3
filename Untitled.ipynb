```python
import pandas as pd
import csv
import glob
import os
os.chdir("Resources")
```

```python
extension = "csv"
bike_data = [x for x in glob.glob('*.{}'.format(extension))]
```

```python
combined_csv = pd.concat([pd.read_csv(f) for f in bike_data])

combined_csv.to_csv("combined_csv.csv", index=False, encoding='utf-8-sig')

combined_csv.head()
```

```python
combined_csv.columns
```

```python
combinedatl = combined_csv[combined_csv["DEST"] == "ATL"]

combinedatl.count()
```

```python
combinedatl['OP_UNIQUE_CARRIER'].unique()
```

```python
combinedatl['OP_UNIQUE_CARRIER'] = combinedatl['OP_UNIQUE_CARRIER'].replace({'DL' :'Delta Airlines',
                                                                         'AA': 'American Airlines',
                                                                         '9E': 'Endeavor Airlines',
                                                                         'WN':'Southwest Airlines',
                                                                         'B6' :'JetBlue Airways',
                                                                         'NK':'Spirit Airlines',
                                                                         'OO':'SkyWest Airlines',
                                                                        'UA': 'United Airlines',
                                                                         'YV':'Mesa Airlines',
                                                                         'YX':'Republic Airline',
                                                                         'AS':'Alaska Airlines',
                                                                         'MQ':'Envoy Air',
                                                                         'OH':'PSA Airlines',
                                                                         'EV':'ExpressJet Airlines',
                                                                         'F9':'Frontier Airlines'
                                                                         })
```
